```python
import pandas as pd 
import pyfaidx
import pysam
import os 
import pathlib
from tqdm import tqdm
from helper_functions import *

refdir = "/Volumes/HNSD01/storage/ref"
outdir = "/Volumes/HNSD01/outdir"
gene_list_version = "v0.1"
output_version = "20250918"

k = 24
additional_flanking_size = 100

path_to_main_output = os.path.join(outdir, "k-mer-feature", "data_analysis", output_version)
path_to_02_output = os.path.join(path_to_main_output, 
                                 "02_output")

os.system(f"mkdir -p {path_to_02_output}")

path_to_hg19 = os.path.join(refdir, "hg19.fa")

all_beds = [item for item in pathlib.Path(f"./custom_tss_gene_beds").glob(f"*/{gene_list_version}/biomart/all_genes/promoter_regions_up_1000_down_1000.bed")]
df = pd.DataFrame()
for file in all_beds:
    tmpdf = pd.read_csv(file, sep="\t", header=None)
    df = pd.concat([df, tmpdf], axis=0)
df.columns = ["chrom", "start", "end", "name"]
df["chrom_num"] = df["chrom"].apply(lambda x: int(x.replace("chr", "")))

df = df.drop_duplicates().reset_index(drop=True)
df.sort_values(by=["chrom_num", "start"], inplace=True)
df = df.drop("chrom_num", axis = 1)
df.to_csv(os.path.join(path_to_02_output, "panel.bed"), sep="\t", index=False, header=False)

os.system(f"mkdir -p ./panels")
df.to_csv(os.path.join("./panels", f"panel_{output_version}.bed"), sep="\t", index=False, header=False)
```
